```python
from langchain_community.document_loaders import ScrapingAntLoader
from bs4 import BeautifulSoup
import pandas as pd

scrapingLoader = ScrapingAntLoader(
    ["https://www.naukri.com/ai-engineer-jobs?k=ai%20engineer"],
    api_key="5506adfd15b5468a816c4d20eb2fdbb4",
    continue_on_failure=True
)
documents = scrapingLoader.load()
```

```python
html = documents[0].page_content
soup = BeautifulSoup(html, "html.parser")

job_cards = soup.find_all("articles")

result = []

```

```python
for card in job_cards:
    try:
        title_tage = card.find("a", class_="title")
        job_title = title_tage.text.strip()
        link = title_tage["href"].strip()

        company = card.find("span", class_="companyName").text.strip()
        location = card.find("li", class_="location").text.strip()

        #Extract job description (short summary insiide <ul>)
        description_item = card.find("ul", class_="job-description")
        if description_item:
            description = " | ".join(li.text.strip() for li in description_item.find_all("li"))
        else:
            description = "N/A"

        result.append({
            "Job Title": job_title,
            "Company" : company,
            "Location": location,
            "Description": description,
            "Link": link
                            })
    except Exception:
        continue
```

```python
df = pd.DataFrame(result)
pd.set_option("display.max_colwidth", None)
print(df.head(10))

```
